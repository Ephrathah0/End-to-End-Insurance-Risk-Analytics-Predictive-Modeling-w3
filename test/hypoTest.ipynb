```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance_data = pd.read_csv('../data/clean_data.csv')
insurance_data.shape
```

```python
insurance_data['ClaimSeverity'] = insurance_data['TotalClaims'].where(insurance_data['TotalClaims'] > 0, np.nan)

claim_data = insurance_data.dropna(subset=['ClaimSeverity'])

provinces = claim_data['Province'].unique()

claim_by_province = [claim_data.loc[claim_data['Province']==prov, 'ClaimSeverity'] for prov in provinces]

f_stat, p_value = stats.f_oneway(*claim_by_province)

print("ANOVA F-statistic:", f_stat)
print("ANOVA p-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: Claim Severity differs significantly across provinces.")
else:
    print("Fail to reject the null hypothesis: No significant difference in Claim Severity across provinces.")
```

```python
zip_claim_freq = pd.crosstab(insurance_data['PostalCode'], insurance_data['has_claim'])
chi2, p, dof, ex = stats.chi2_contingency(zip_claim_freq)
print("Chi-squared test for claim frequency by zip code")
print("Chi2:", chi2, "p-value:", p)
if p < 0.05:
    print("Reject H0: Claim frequency differs by zip code")
else:
    print("Fail to reject H0: No significant difference in claim frequency by zip code")

zip_with_claims = insurance_data.dropna(subset=['ClaimSeverity'])
zip_codes = zip_with_claims['PostalCode'].unique()
claim_by_zip = [zip_with_claims.loc[zip_with_claims['PostalCode']==z, 'ClaimSeverity'] for z in zip_codes]
f_stat_zip, p_zip = stats.f_oneway(*claim_by_zip)
print("ANOVA for Claim Severity by zip code")
print("F-statistic:", f_stat_zip, "p-value:", p_zip)
if p_zip < 0.05:
    print("Reject H0: Claim Severity differs by zip code")
else:
    print("Fail to reject H0: No significant difference in claim severity by zip code")
```

```python
import scipy.stats as stats

# Margin = TotalPremium - TotalClaims
insurance_data['Margin'] = insurance_data['TotalPremium'] - insurance_data['TotalClaims']

# Prepare data for ANOVA
zip_codes = insurance_data['PostalCode'].unique()
margin_by_zip = [insurance_data.loc[insurance_data['PostalCode']==z, 'Margin'] for z in zip_codes]

# Run ANOVA
f_stat_margin, p_val_margin = stats.f_oneway(*margin_by_zip)
print("ANOVA for Margin by zip code")
print("F-statistic:", f_stat_margin, "p-value:", p_val_margin)

if p_val_margin < 0.05:
    print("Reject H0: Margin differs by zip code")
else:
    print("Fail to reject H0: No significant difference in margin by zip code")
```
